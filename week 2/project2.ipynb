```python
import numpy as np

print("What operation would you like to perform?")
print("1. Root of the cubic equation (Ax³ + Bx² + Cx + D = 0)")
print("2. The root of the quartic equation (Ax⁴ + Bx³ + Cx² + Dx + E = 0)")
print("3. The root of a quadratic equation (Ax² + Bx + C = 0)")

choice = input("Choose an input (1-3): ")

# Case 1: Cubic Equation
if choice == "1":
    print("\n--- Solving Cubic Equation ---")
    a = float(input("Enter coefficient A: "))
    b = float(input("Enter coefficient B: "))
    c = float(input("Enter coefficient C: "))
    d = float(input("Enter coefficient D (constant): "))

    cubic_coeffs = [a, b, c, d]
    roots = np.roots(cubic_coeffs)
    print("Roots:", roots)

# Case 2: Quartic Equation
elif choice == "2":
    print("\n--- Solving Quartic Equation ---")
    a = float(input("Enter coefficient A: "))
    b = float(input("Enter coefficient B: "))
    c = float(input("Enter coefficient C: "))
    d = float(input("Enter coefficient D: "))
    e = float(input("Enter coefficient E (constant): "))

    quartic_coeffs = [a, b, c, d, e]
    roots = np.roots(quartic_coeffs)
    print("Roots:", roots)

# Case 3: Quadratic Equation
elif choice == "3":
    print("\n--- Solving Quadratic Equation ---")
    a = float(input("Enter coefficient A: "))
    b = float(input("Enter coefficient B: "))
    c = float(input("Enter coefficient C (constant): "))

    # Calculating the Discriminant
    discriminant = (b**2) - (4*a*c)

    if discriminant > 0:
        r1 = (-b + discriminant**0.5) / (2*a)
        r2 = (-b - discriminant**0.5) / (2*a)
        print(f"Two real roots: {r1} and {r2}")
    elif discriminant == 0:
        r = -b / (2*a)
        print(f"One real root: {r}")
    else:
        # Handling complex roots using Python's complex number support
        r1 = (-b + (complex(discriminant)**0.5)) / (2*a)
        r2 = (-b - (complex(discriminant)**0.5)) / (2*a)
        print(f"Complex roots: {r1} and {r2}")

else:
    print("Invalid choice. Please run the program again.")
```
