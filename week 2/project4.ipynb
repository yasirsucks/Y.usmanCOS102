```python
# ATR calculator for Izifin Technology
print("Hello! and welcome to Izifin Technology")
print("All fields must be filled with numerical years.")

# Convert inputs to integers
xp = int(input("What is your work experience (years)? "))
age = int(input("How old are you? "))

# Logic for Annual Tax Revenue (ATR)
if xp > 25 and age >= 55:
    atr = "N5,600,000"
elif xp > 20 and age >= 45:
    atr = "N4,480,000"
elif xp > 10 and age >= 35:
    atr = "N1,500,000"
else:
    # This covers staff with less than 10 years experience 
    # and those who don't fit the higher brackets
    atr = "N550,000"

print(f"\nResults for staff:")
print(f"Experience: {xp} years")
print(f"Age: {age} years")
print(f"Your Annual Tax Revenue (ATR) is: {atr}")
```
